```python
%matplotlib qt5
import itfit
import numpy as np
from tests import dataFunction, gauss
import matplotlib.pyplot as plt
```

```python
noise = np.random.normal(size=200)

xdata = np.arange(200)
ydata = dataFunction(xdata, -0.04, 5, np.random.random()
                     * 30, np.random.random()*200, 15) + noise
```

```python
fitter_app = itfit.Fitter(xdata, ydata)
fitter_app()
```

```python
fit = fitter_app.get_plot_builder().style("science")
fit.plot_fit(label="fit").with_data(label="data")
fit.grid().legend().tight_layout()
```

```python
fit.labels().start_title("This may be a bad example").color("red").end_title().end_labels()#.show_inline()
```

```python
fit.labels().start_x_label("time [s]").fontsize(12).end_xlabel()\
    .start_y_label("current [A]").fontsize(12).end_ylabel()\
        .end_labels()
        # .show_inline()
```

```python
fit.set_size((12,9)).tight_layout()
```

```python
fit.tight_layout().save_fig("pato.svg").show_inline()
```
